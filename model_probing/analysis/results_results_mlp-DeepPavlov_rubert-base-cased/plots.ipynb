```python
from matplotlib import pyplot as plt

def plot_each_column_in_same_plot(df):
    # plot each column in same plot
    df.plot()
    plt.show()
```

```python

def plot_list_of_columns_in_y_axis(df, columns, plot_title='Precision'):
    # plot each column in same plot

    df.plot(y=columns)
    plt.title(plot_title)
    plt.legend(['Acc', 'Dat', 'Gen', 'Loc', 'Ins', 'Nom'], loc='lower right')
    plt.show()

```
