```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
from matplotlib import pyplot as plt

def plot_each_column_in_same_plot(df):
    # plot each column in same plot
    df.plot()
    plt.show()
```

```python

def plot_list_of_columns_in_y_axis(df, columns, plot_title='Precision'):
    # plot each column in same plot

    df.plot(y=columns)
    plt.title(plot_title)
    plt.legend(['Acc', 'Dat', 'Gen', 'Loc', 'Ins', 'Nom'], loc='lower right')
    plt.show()

```

```python
plot_list_of_columns_in_y_axis(df,
                               columns=['0_precision', '1_precision', '2_precision', '3_precision', '4_precision', '5_precision'])
```

```python
plot_list_of_columns_in_y_axis(df,
                               columns=['0_recall', '1_recall', '2_recall', '3_recall', '4_recall', '5_recall'],
                               plot_title='Recall')
```
